```python
import matplotlib.pyplot as plt
a=1
while(a):
    try:
        radius=int(round(float(input("Enter the radius:"))))
        x_center,y_center=(round(float(i)) for i in input("Enter the starting point(x y):").split())
        if radius > 0:
            a = 0
        else:
            print("Need to enter positive value.")

    except:
            print('invalid input')
            print('enter again')
print(radius)
```

```python
def second_octant(radius):
    P0=1-radius  #initial parameter 
    Pk=P0
    x=0
    y=radius
    points=[[x,y]]  #[x,y]
    while x<y:
        x+=1
        if Pk<0:  #y increment =0  [Xk+1,Yk]       x=Xk , y=Yk
            Pk=Pk+2*x+1
        else:   #y increment =-1
            y -= 1
            Pk = Pk + 2 * (x - y) + 1
        points.append([x,y])  #[X(k+1),Y(k+1)]
    return points
```

```python
oct_2=second_octant(radius)
```

```python
def first_quad(oct_2):
    oct_1=[]
    length=len(oct_2)-2
    for i in range(length):
        oct_1.append([oct_2[length-i-1][1],oct_2[length-i-1][0]]) #[x,-y]
    return oct_2+oct_1
        

```

```python
quad_1=first_quad(oct_2)
```

```python
def semi_circle(quad_1):
    length=len(quad_1)-2
    for i in range(length+1):
        quad_1.append([-quad_1[i+1][0],quad_1[i+1][1]]) #[-x,y]
    return quad_1
```

```python
semi_1=semi_circle(quad_1)
print(semi_1)
```

```python
def circle(semi_1):
    length=len(semi_1)-2
    for i in range(length+1):
        semi_1.append([semi_1[i][0],-semi_1[i][1]])  #[x,-y]
    return semi_1
```

```python
circle(semi_1)
unique_list = list(set(tuple(point) for point in circle(semi_1)))
unique_list = [list(point) for point in unique_list]
print (unique_list)
for i in unique_list:
    i[0]=i[0]+x_center
    i[1]=i[1]+y_center
```
