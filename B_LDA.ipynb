```python
import matplotlib.pyplot as plt
a=1
while(a):
    try:
        point1=[round(float(i)) for i in input("Enter the starting point(x y):").split()]
        point2=[round(float(i)) for i in input("Enter the ending point(x y):").split()]
        a=0
        if len(point1) == 2 and len(point2) == 2:
            a = 0
        else:
            print("Need to enter exactly x and y coordinates (2 numbers).")

    except:
            print('invalid input')
            print('enter again')
```

```python
dy=abs(point2[1]-point1[1])
dx=abs(point2[0]-point1[0])
sx= 1 if point2[0] > point1[0] else -1
sy=1 if point2[1] > point1[1] else -1
def case1(point1,point2):    #for dx>dy
   x=[point1[0]]
   Pk=2*dy-dx
   y=[point1[1]]
   for z in range(0,dx):
        if (Pk<0):
              Pk=Pk+2*dy
              y.append(y[-1])
        else:
              Pk=2*dy-2*dx+Pk
              y.append(y[-1]+sy)
        x.append(x[-1]+sx)
   return x,y
```

```python
def case2(point1,point2):    #for dy>dx
   y=[point1[1]]
   Pk=2*dx-dy
   x=[point1[0]]
   for z in range(0,dy):
          if (Pk<0):
              Pk=Pk+2*dx
              x.append(x[-1])
          else:
              Pk=2*dx-2*dy+Pk
              x.append(x[-1]+sx)
          y.append(y[-1]+sy)
   return x,y
```

```python
if (dx>dy):
   x,y=case1(point1,point2)
else:
   x,y=case2(point1,point2)
plt.scatter(x, y, color='red') # Plot the pixels
plt.plot(x, y, color='blue', alpha=0.3) # Draw a light line through them
plt.xticks(range(min(x)-1, max(x) + 2))
plt.yticks(range(min(y)-1, max(y) + 2))
plt.title("Bresenham's Line Drawing Algorithm")
plt.grid(True)
plt.show()
points=[[x[i],y[i]] for i in range (len(x)-1)]
print(points)
```
